```python
import pandas as pd

df = pd.read_csv("./qualitaet_spritzguss.csv")
df.head()
```

```python
#matplotlib inline
import matplotlib.pyplot as plt
plt.set_cmap("bwr")

plt.scatter(df["Druck"], df["Temperatur"], c = df["Erfolg"])
plt.xlabel("Druck [bar]")
plt.ylabel("Temperatur [Â°C]")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Welche Spalten sollen zur Vorhersage verwendet werden
X = df[["Druck", "Temperatur"]].values
y = df["Erfolg"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

```python
# Testdaten plotten

plot_classifier(model, X_test, y_test, proba = False, xlabel = "Druck", ylabel = "Temperatur", cmapName = 'bwr')
```
